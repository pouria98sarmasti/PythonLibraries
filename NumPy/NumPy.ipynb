```python
# calculate execution time using 2 method

import time
from timeit import Timer
import numpy as np

##### method1: time module
vector_size = 1000000

li1 = list(range(vector_size))
li2 = list(range(vector_size))
# function for time of a sum operation on list
def time_li():
    t1 = time.time()
    li3 = [li1[i]+li2[i] for i in range(vector_size)]
    return time.time() - t1

arr1 = np.arange(vector_size)
arr2 = np.arange(vector_size)
# function for time of a sum operation on array
def time_arr():
    t1 = time.time()
    arr3 = arr1 + arr2
    return time.time() - t1

# now compare sum operation time
print("method1:\n","list:", time_li())
print("array:", time_arr())
print("list/arr:", time_li()/time_arr(),"\n\n")


##### method2: timeit module
# function for sum on list
def timer_li():
    li3 = [li1[i]+li2[i] for i in range(vector_size)]

# function for sum of array
def timer_arr():
    arr3 = arr1 + arr2

# creating timer object
timer_li_obj = Timer("timer_li","from __main__ import timer_li")
timer_arr_obj = Timer("timer_arr","from __main__ import timer_arr")

# comaprsion time of sum operasion
print("method2:\n","list:", timer_li_obj.timeit(10))
print("array:", timer_arr_obj.timeit(10))
print("list/arr:", timer_li_obj.timeit(10)/timer_arr_obj.timeit(10),"\n\n")
```

27. working with missing values
    1. provide converters for colomns that have missing values 
        1. converters: {3: func3, 4: func4}
        2. converter: assign toxic value instead of missing values
    2. load data: arr1 = np.loadtxt(converters)
    3. select aviable values: aviable_weight = arr1["weight"] > 0

    

```python
import numpy as np

# create an array from a function
arr1 = np.fromfunction(lambda i,j: i==j, (3,3), dtype=int)
print(arr1)

# from a iter
it = (i**2 for i in range(6))
arr1 = np.fromiter(it, dtype=int, count=3)
print(arr1)

# from string
name = '1 2 3 45'
arr1 = np.fromstring(name, dtype=int, sep=' ')
print(arr1)

# !load text, and np.dtype
file_name = "file1.txt"
dtype1 = np.dtype([("sex", "S2"), ("height", "f2")])
arr2 = np.loadtxt(file_name, dtype=dtype1, skiprows=9, usecols=(1, 3))
print(arr2)

# access data from a loaded text file #! important
male = arr2["height"] == b"M"

# extract special data of a loaded text #! important
avg_male_height = arr2["height"][male].mean()
avg_fmale_height = arr2["height"][~male].mean()

# working with weight colomn that has missing values #! missing values in numpy
def correct_missing_weights(weight):
    try:
        return float(weight)
    except ValueError:
        return float(-100)

dtype2 = np.dtype([("sex", "S2"), ("weight", "f2")])
arr3 = np.loadtxt(file_name, dtype=dtype2, skiprows=9, usecols=(1, 4), converters={4:correct_missing_weights})
aviable_weights = arr3["weight"] > 0
avg_male_weight = arr3["weight"][aviable_weights & male].mean()
avg_fmale_weight = arr3["weight"][aviable_weights & ~male].mean()

```

37. access specific data in an array
    1. specific rows: arr1[[2, 3, 4]]
    2. specific cols: arr1[:, [2, 3, 4]]
    3. specific elements: arr1[[2, 3, 4], [2, 3, 4]]
    4. specific rows and cols: arr1[[2, 3, 4]][:,[2, 3, 4]]
    5. this indexing create a copy not view

```python
import numpy as np

# create data and name #! np.random.randint
names = np.array(['pouria', 'nastaran', 'parisa', 'payam', 'ali', 'mohtaram', 'ali'])
data = np.random.randint(0, 10, size=(7,4))

# create boolean index for rows
row_bool = names == 'pouria'
# boolean indexing #! return copy
pouria = data[row_bool, :3]
pouria[0,0] = 125
print(data, pouria)
print(data[~row_bool])

# specific data in an array
arr1 = np.arange(25).reshape(5, 5)
print("\n\n\n", arr1)
# rows
print("rows:\n",arr1[[0, 2]])
# cols
print("cols:\n",arr1[:,[0, 2]])
# elements
print("elements:\n", arr1[[0, 2], [0, 2]])
# rows and cols
print("rows and cols:\n", arr1[[0, 2]][:, [0, 2]])

# copy test #! fancy indexing also create copy
arr2 = arr1[[1]][:,[2]]
print(arr2.shape)
arr2[0, 0] = -10
print(arr1, arr2)
```
